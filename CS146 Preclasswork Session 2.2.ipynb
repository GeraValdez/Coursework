```python
import numpy as np
import scipy.stats as stats 
from matplotlib import pyplot as plt

import seaborn as sns

x = np.linspace (0, 100, 100) 
```

```python
#Normal Distribution 1: mean equals 15, std equals 3
y1 = stats.norm.pdf(x,15,3)

#Normal Distribution 2: mean equals 25, std equals 4
y2 = stats.norm.pdf(x,25,4)

#Graph
plt.plot(x, y1, 'r',"y-", label=(r'$\bar{x}=15, \sigma=3$'))
plt.plot(x, y2, 'g',"y-", label=(r'$\bar{x}=25, \sigma=4$'))
plt.ylim([0,0.20])
plt.xlim([0,50])
plt.title('Comparing two normal distributions')
plt.xlabel('Width (units)')
plt.ylabel('Height (units)')
plt.legend()
plt.show()
```

```python
#Gamma Distribution 1: alpha equals 35, beta equals 3
y1 = stats.gamma.pdf(x, a=35, scale=0.33333)

#Gamma Distribution 2: alpha equals 25, beta equals 3
y2 = stats.gamma.pdf(x, a=25, scale=0.33333)

#Graph 
plt.plot(x, y1, 'r',"y-", label=(r'$\alpha=35, \beta=3$')) 
plt.plot(x, y2, 'g',"y-", label=(r'$\alpha=25, \beta=3$')) 
plt.ylim([0,0.3])
plt.xlim([0,40])
plt.title('Comparing two alpha distributions')
plt.xlabel('Width (units)')
plt.ylabel('Height (units)')
plt.legend()
plt.show()
```

```python
#Beta Distribution 1: alpha equals 2, beta equals 4
y1 = stats.beta.pdf(x,2.0,4.0, scale=100, loc=0)

#Beta Distribution 2: alpha equals 3, beta equals 2
y2 = stats.beta.pdf(x,3.0,2.0, scale=100, loc=0)

#Graph 
plt.plot(x, y1, 'r',"y-", label=(r'$\alpha=35, \beta=3$')) 
plt.plot(x, y2, 'g',"y-", label=(r'$\alpha=25, \beta=3$')) 
plt.ylim([0,0.1])
plt.xlim([0,100])
plt.title('Comparing two beta distributions')
plt.xlabel('Width (units)')
plt.ylabel('Height (units)')
plt.legend()
plt.show()
```
